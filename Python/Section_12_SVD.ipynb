## SVD and Eigandecomposition for Square Symmetric Matrices

```python
# Determine the relationship between eig and svd for symmetric matrices
# Create a 5x5 symmetric matrix
A = np.random.randn(5,5)
A = A.T@A

# Compute eig and svd
L,W = np.linalg.eig(A)
print(np.diag(L)) # LAMBDA
print()
print(W) # Eiganvectors
```

```python
U,S,V = np.linalg.svd(A)
print(U) # Vector U
print()
print(np.diag(S)) # Lambda values
print()
print(V) # Vector S
```

## Get U from Eigandecomposition of $A^TA$

```python
# Create a rectangular, wide matrix
m = 3
n = 6
A = np.random.randn(m,n)

# Full svd() (Name: Us, Ss, Vs)
Us, Ss, Vs = np.linalg.svd(A)

# Eigandecomposition of A.TA
L,V = np.linalg.eig(A.T@A)

# Sort the eigansolution for comparison
sidx = np.argsort(L)[::-1]
L = L[sidx]
V = V[:,sidx]
```

```python
# Part 1: Confirm Vs = V
print(np.round(Vs.T - V,2))
print()
print(np.round(Vs.T + V,2))
# only first three columns explained because A is rank 3
```

```python
# Part 2: Check the relationship between Ss and L
print(L[0:3]) # The rest are zero
print()
print(Ss**2)
```

```python
# Part 3: Create U using A,V and L
U = np.zeros((m,m))

for i in range(m):
    U[:,i] = A@V[:,i].T/np.sqrt(L[i])

print(U)
```

```python
# Because of sign differences, + and - complement each other
print(np.round(Us - U,2))
print()
print(np.round(Us + U,2))
```

## Spectral Theory of Matrices

```python
# Define matrix dimensions
m = 40
n = 30

# Define a 2D Gaussian for smoothing
k = int((m+n)/4)
xx = np.linspace(-3,3,k)
[X,Y] = np.meshgrid(xx,xx)
g2d = np.exp(-(X**2 + Y**2)/(k/2))

# Define a matrix
A = signal.convolve2d(np.random.randn(m,n), g2d, "same")
plt.imshow(A)
plt.show()
```

```python
# Remember, V is transposed by efault in Python
U,S,V = np.linalg.svd(A)

# Show constituent matrices
fig = plt.subplots(1,figsize=(10,6))

plt.subplot(241)
plt.imshow(A)
plt.title('A')

plt.subplot(242)
plt.imshow(U)
plt.title('U')

plt.subplot(243)
plt.imshow(np.diag(S))
plt.title('$\Sigma$')

plt.subplot(244)
plt.imshow(V)
plt.title('V$^T$')

plt.subplot(212)
plt.plot(S,'ks-')
plt.xlabel('Component number')
plt.ylabel('$\sigma$')
plt.title('"Scree plot" of singular values')

plt.show()
```

```python
# One layer at a time
rank1mats = np.zeros((6,m,n))

for i in range(0,6):
    
    # create rank1 matrix
    rank1mats[i,:,:] = np.outer( U[:,i]*S[i],V[i,:] )
    
    plt.subplot2grid((2,6),(0,i))
    plt.imshow(rank1mats[i,:,:],vmin=-5,vmax=5)
    plt.axis('off')
    plt.title('C.%d' %(i+1))
    
    plt.subplot2grid((2,6),(1,i))
    imdat = np.sum(rank1mats[:i+1,:,:],axis=0)
    plt.imshow(imdat,vmin=-5,vmax=5)
    plt.axis('off')
    plt.title('Cs 1:%d' %(i+1))

plt.show()

# then show the original
plt.imshow(A,vmin=-5,vmax=5)
plt.show()
```

## SVD for Low-Rank Approximation

```python
# The code in this section depends on the one above
# Define tbe number of components you wanna keep
nComps = 5

# Define reduced matrices
Ur = U[:,0:nComps]
Sr = S[0:nComps]
Vr = V[0:nComps,:]

# Lower rank approximation
reconImage = Ur@np.diag(Sr)@Vr

# Confirm rank as 5
print("The rank of Reconstructed Matrix is {}".format(np.linalg.matrix_rank(reconImage)))
```

```python
# Error map and percent difference
errormap = (reconImage - A)**2

# Create images
plt.figure(figsize=(10,10))
plt.subplot(131)
plt.imshow(A,vmin=-10, vmax=10)
plt.axis("off")
plt.title("Matrix A")

plt.subplot(132)
plt.imshow(reconImage, vmin=-10,vmax=10)
plt.axis("off")
plt.title("Recon Image")

plt.subplot(133)
plt.imshow(errormap, vmin=-10, vmax=10)
plt.axis("off")
plt.title("Errormap")

plt.show()
```

```python
# overall percent error
100*np.linalg.norm(reconImage - A)/np.linalg.norm(A)
```

## SVD, Matrix Inverse and Pseudoinverse

```python
# Define the matrix
A = np.array([[1,2,3],
              [1,2,4], # change 2->0 for invertible matrix later
              [1,2,5]])

# SVD
U,S,V = np.linalg.svd(A)

# Pseudoinvert
nonzeroels = S>10**(-14)
S[nonzeroels] = 1/S[nonzeroels]

Ai = V.T@np.diag(S)@U.T

# Check whether its close to I
print(Ai@A)
print()

# Compare to pinv() function
print(np.linalg.pinv(A)@A)
# They are the same
```

## Create Matrix with Desired Condition Number

```python
# Create a matrix with a specific conditional number
m = 6
n = 16

condnum = 42

# create singular matrices: gotta be orthogonal
U,J = np.linalg.qr(np.random.randn(m,m))
V,J = np.linalg.qr(np.random.randn(n,n))

# Create the singular value matrix
s = np.linspace(condnum,1,min(m,n))

S = np.zeros((m,n))
for i in range(len(s)):
    S[i,i] = s[i]

# Create the ,atrix using SVD Formula
A = U@S@V.T
condA = np.linalg.cond(A)
print(condA)
```

```python
fig = plt.subplots(1,figsize=(8,5))
plt.subplot(231)
plt.imshow(U)
plt.axis("off")
plt.title("U")

plt.subplot(232)
plt.imshow(S)
plt.axis("off")
plt.title("$\Sigma$")

plt.subplot(233)
plt.imshow(V.T)
plt.axis("off")
plt.title("$V^T$")

plt.subplot(235)
plt.imshow(A)
plt.axis("off")
plt.title("A, Cond=%g"%condA)

plt.show()
```

## Why Avoid Matrix Inverse

```python
# Create a matrix with a known condition number
# compute its explicit inverse using inv() function
# Multiply the two to get I
# Compute the norm of the difference between I and eye()
# Repeat for matrices of size 2-70 and condition numbers btw 10-10^12
# Show results in an image

matrixSizes = np.arange(2,71)
condnums = np.linspace(10,1e12,40)

# define value differences matrix
invDiffs = np.zeros((len(matrixSizes), len(condnums)))

# Create the for loop
for mi,M in enumerate(matrixSizes):
    for ci,C in enumerate(condnums):
        
        # Create matrix A from SVD
        U = np.linalg.qr(np.random.randn(M,M))[0]
        V = np.linalg.qr(np.random.randn(M,M))[0]
        S = np.diag(np.linspace(C,1,M))
        A = U@S@V
        
        # Difference from identity matrix
        I = A@np.linalg.inv(A)
        idiff = np.abs(I - np.eye(M))
        
        # Get the norm
        invDiffs[mi,ci] = np.linalg.norm(idiff,2)
        
# Visualize the whole thing
fig = plt.figure(figsize = (10,10))
plt.pcolor(condnums,matrixSizes,invDiffs,vmax=np.max(invDiffs)*0.6,shading="auto")
plt.xlabel("Condition Number")
plt.ylabel("Matrix Size")
plt.colorbar()
plt.show()
```
