# Chapter 5 - Matrix Spaces

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Column Space of a Matrix

```python
# Define a matrix
S = np.array([[3,0],
              [5,2],
              [1,2]])

# define a vector within space
v = np.array([-3, 1, 5]).T
# v = np.array([[1], [7], [3]])

plt.figure(figsize=(10, 10))
#fig = plt.figure()
ax = plt.axes(projection = "3d")

# draw plane corrsponding to the column space
xx, yy = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 10))
cp = np.cross(S[:,0], S[:,1])
z1 = (-cp[0]*xx - cp[1]*yy)/cp[2] # normal vectors
ax.plot_surface(xx,yy,z1,alpha = 0.2)

# Plot the two vectors from matrix S
ax.plot([0, S[0,0]], [0, S[1,0]], [0, S[2,0]], "k")
ax.plot([0, S[0,1]], [0, S[1,1]], [0, S[2,1]], "k")

# and vector v
ax.plot([0, v[0]], [0, v[1]], [0, v[2]], "r")

ax.view_init(elev=155, azim=0)
plt.show()
```
