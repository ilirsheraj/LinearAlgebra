## Finding Eiganvalues

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define 2x2 matrix
A = [ [1,5], [2,4]]

# extract the eiganvalues
eigvals = np.linalg.eig(A)
# first element is eiganvalu and second one is eiganvector
print(eigvals[0])
print()
print(eigvals[1])
```

```python
# define two vectors
# eiganvector of Matrix A
v1 = np.array([1,1])

# a random vector, quite unlikely to be eiganvector of A
v2 = np.random.randn(2,1)
# take its unit length
v2 = v2/np.linalg.norm(v2)

# compute Av
Av1 = A@v1
Av2 = A@v2

print(Av1)
print()
print(Av2)
```

```python
# Plot the vectors and Av
plt.plot([0,v1[0]], [0,v1[1]], "r")
plt.plot([0,Av1[0]], [0,Av1[1]], "r--")
plt.plot([0,v2[0]], [0,v2[1]], "k")
plt.plot([0,Av2[0]], [0,Av2[1]], "k--")
plt.axis([-8,8,-8,8])
plt.grid()
plt.legend(["v1", "Av1", "v2", "Av2"])
plt.show()
```

```python
# Change the matrix to see what happens when the vector is not eiganvector
A = [ [1,5], [2,1]]

# original eiganvector
v1 = np.array([1,1])

# a random vector, quite unlikely to be eiganvector of A
v2 = np.random.randn(2,1)

# take its unit length
v2 = v2/np.linalg.norm(v2)

# compute Av
Av1 = A@v1
Av2 = A@v2

# Plot the vectors and Av
plt.plot([0,v1[0]], [0,v1[1]], "r")
plt.plot([0,Av1[0]], [0,Av1[1]], "r--")
plt.plot([0,v2[0]], [0,v2[1]], "k")
plt.plot([0,Av2[0]], [0,Av2[1]], "k--")
plt.axis([-8,8,-8,8])
plt.grid()
plt.legend(["v1", "Av1", "v2", "Av2"])
plt.show()
```

```python
# Confirm the eiganvalues for the 3x3 matrix
A = np.array([[-2,2,-3],
              [-4,1,-6],
              [-1,-2,0]])

evals = np.linalg.eig(A)
print(evals[0])
```

## Diagonalization 

```python
# create a symmetric matrix
A = np.round(10*np.random.randn(4,4))
A = A.T@A

# eigandecomposition
evals, evecs = np.linalg.eig(A)

# reconstruct
Ap = evecs @ np.diag(evals) @ np.linalg.inv(evecs)

plt.subplot(121)
plt.imshow(A)
plt.axis("off")
plt.title("A")

plt.subplot(122)
plt.imshow(Ap)
plt.axis("off")
plt.title("$VAV^{-1}$")

plt.show()
```

```python
# Difference in RMS
recondiff = A - Ap
rmsA = np.sqrt(np.mean(np.square(np.reshape(recondiff, (1,-1)))))
print("Reconstruction RMS:", rmsA)
```

```python
# Visualize Diagonalization
A = np.round(10*np.random.randn(10,10))
A = A.T@A

# eigandecomposition
D,V = np.linalg.eig(A)

plt.subplot(141)
plt.imshow(A)
plt.axis("off")
plt.title("A")

plt.subplot(142)
plt.imshow(V)
plt.axis("off")
plt.title("V")

plt.subplot(143)
plt.imshow(np.diag(D))
plt.axis("off")
plt.title("$\Lambda$")

plt.subplot(144)
plt.imshow(np.linalg.inv(V))
plt.axis("off")
plt.title("$V^{-1}$")

plt.show()
# A = V*Lambda*V-inverse
```

## Matrix Powers via Diagonalization

```python
# matrix power: 3
A = np.random.randn(2,2)

# direct matrix power
A3 = np.linalg.matrix_power(A,3)
A31 = A@A@A

# eigandecomposition
D,V = np.linalg.eig(A)
D = np.diag(D)

A3e = V @ np.linalg.matrix_power(D,3) @ np.linalg.inv(V)

print(A3)
print()
print(A31)
print()
print(A3e)
# All are the same
```

```python
# eigenvalues of A and A^3
# eiganvectors are the same, eiganvalues are cubed
A = np.random.rand(3,3)
A = A@A.T

D,V   = np.linalg.eig(A)
D3,V3 = np.linalg.eig( A@A@A )

print(V) 
print()
print(V3)

plt.subplot(121)
plt.imshow(V)
plt.axis('off')
plt.title('evecs of A')

plt.subplot(122)
plt.imshow(V3)
plt.axis('off')
plt.title('evecs of A^3')

plt.show()
```

## Eigandecomposition of Matrix Differences

```python
# Define two random matrices and make them symmetric
A = np.random.randn(5,5)
A = A.T@A

B = np.random.randn(5,5)
B = B.T@B

D1,V1 = np.linalg.eig(A-B)
D2,V2 = np.linalg.eig(A@A - A@B - B@A + B@B)

print(np.round(D1,3))
print()
print(np.round(D2,3))
print()
print(np.round(D1**2,3))
```

```python
# Sort the Eiganvalues
sidx1 = np.argsort(abs(D1))
sidx2 = np.argsort(abs(D2))
print(sidx1)
print()
print(sidx2)
```

```python
# sort eiganvalues and vectors
D1 = D1[sidx1]
D2 = D2[sidx2]
print(D1**2)
print()
print(D2)
```

```python
v1 = V1[:,sidx1]
v2 = V2[:,sidx2]

np.round(v1-v2,3)
```

```python
# the complement here
np.round(v1+v2,3)
```

```python
# Sign flips in some places, otherwise its same
print(np.round(v1,2))
print()
print(np.round(v2,2))
```

```python
# If we dont use symmmetric matrices
A = np.random.randn(5,5)
#A = A.T@A

B = np.random.randn(5,5)
#B = B.T@B

D1,V1 = np.linalg.eig(A-B)
D2,V2 = np.linalg.eig(A@A - A@B - B@A + B@B)

print(np.round(D1,3))
print()
print(np.round(D2,3))
print()
print(np.round(D1**2,3))
```

## Reconstruct Matrix from Eiganlayers

```python
# create a symmetric matrix and carry out eigandecomposition
m = 5
A = np.random.randn(5,5)
A = np.round(10*A.T@A)

D,V = np.linalg.eig(A)

# Convert to column vector
v = V[:,2]
# impose orientation on the vector to turn it to column vector
print(np.reshape(V[:,2],(m,1)))
print()
print(np.reshape(V[:,2],(m,1)).T)
```

```python
# Show that the norm of the outer product of v_i is 1\
# no need for orientation
op = np.outer(v,v)
print(op)
```

```python
# We can however impose orientation
v = np.reshape(V[:,2],(m,1))
print(v*v.T) # With oreintation we can use *
print()
print(v@v.T) # normal way to get outer product
print()
print(np.linalg.norm(v)) # prove that the norm is indeed 1
```

```python
# create one layer of A as $\lambda$*v*v
# Norm of outer product times eigabvalue is equal to eiganvalue itself
print(np.linalg.norm(v*D[2]*v.T))
print(D[2])
```

```python
# Reconstruct A by summing over the outer products
Arecon = np.zeros((m,m))

for i in range(m):
    v = np.reshape(V[:,i], (m,1))
    Arecon += v*D[i]*v.T
    print(Arecon)
    print("Matrix Rank is {}".format(np.linalg.matrix_rank(Arecon)))
    print()

print()
print(A)
print()
print(Arecon - A)
```

## Trace and Determinant, Eiganvalues Sum & Product

```python
# trace(A) = sum(D)
# det(A) = prod(D)
# First for full-rank matrix
A = np.random.randn(7,7)
tr = np.matrix.trace(A)
dt = np.linalg.det(A)
l = np.linalg.eig(A)[0]

# They must be the same
print(tr, sum(l))
print()
print(dt, np.prod(l))
```

```python
# Reduced Rank Matrix
A = np.random.randn(7,5) @ np.random.randn(5,7)
tr = np.matrix.trace(A)
dt = np.linalg.det(A)
l = np.linalg.eig(A)[0]

# They must be the same
print(tr, sum(l))
print()
# Determinant is zero because min 2 eiganvalues are zero
print(dt, np.prod(l))
```

## Generalized Eigandecomposition

```python
import scipy.linalg # since numpy doenst do GED
```

```python
# Define two matrices
A = np.array([[3,2], [1,3]])
B = np.array([[1,1], [4,1]])

# Carry out GED
eigvals,eigvecs = scipy.linalg.eig(A,B)

print(eigvals)
print()
print(eigvecs)
```

```python
# Matrix-Vector Multiplication
# we can use any eiganvector, but it mst be consistent
Av = A@eigvecs[:,1]
Bv = B@eigvecs[:,1]
BinvAv = np.linalg.inv(B)@A@eigvecs[:,1]
```

```python
plt.plot([0,eigvecs[0,1]],[0,eigvecs[1,1]],'k',linewidth=4,label='$v_2$')
plt.plot([0,Av[0]],[0,Av[1]],'r--',linewidth=2,label='$Av_2$')
plt.xlim([-3,3]), plt.ylim([-3,3])
plt.plot([-3,3],[0,0],'k:')
plt.plot([0,0],[-3,3],'k:')
plt.legend()
plt.title('Av')
plt.show()

plt.plot([0,eigvecs[0,1]],[0,eigvecs[1,1]],'k',linewidth=4,label='$v_2$')
plt.plot([0,Bv[0]],[0,Bv[1]],'r--',linewidth=2,label='$Bv_2$')
plt.xlim([-3,3]), plt.ylim([-3,3])
plt.plot([-3,3],[0,0],'k:')
plt.plot([0,0],[-3,3],'k:')
plt.legend()
plt.title('Bv')
plt.show()


plt.plot([0,eigvecs[0,1]],[0,eigvecs[1,1]],'k',linewidth=4,label='$v_2$')
plt.plot([0,BinvAv[0]],[0,BinvAv[1]],'r--',linewidth=2,label='B$^{-1}$Av_2')
plt.xlim([-3,3]), plt.ylim([-3,3])
plt.plot([-3,3],[0,0],'k:')
plt.plot([0,0],[-3,3],'k:')
plt.legend()
plt.title('B$^{-1}$Av')
plt.show()
```

## GED in Small and Large Matrices

```python
# Compare eig(S,R) to eig(inv(R)*S)
S = np.random.randn(2,2)
R = np.random.randn(2,2)

Ls,Ws = scipy.linalg.eig(S,R)
Li,Wi = scipy.linalg.eig(np.linalg.inv(R)@S)

print(Ls)
print()
print(Li)
```

```python
# Lets plot the eiganvectors
plt.plot([0,Ws[0,0]],[0,Ws[1,0]],"k")
plt.plot([0,Ws[0,1]],[0,Ws[1,1]],"k--")
plt.plot([0,Wi[0,0]],[0,Wi[1,0]],"r")
plt.plot([0,Wi[0,1]],[0,Wi[1,1]],"r--")

plt.axis("square")
plt.grid()
#plt.axis([-1, 1, -1, 1])
plt.show()
# sign flips may happen, but they are identical
```
