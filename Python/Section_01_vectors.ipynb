## Dot Product Properties

```python
# Distributive Property
n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

res1 = np.dot(a, (b + c))
res2 = np.dot(a,b) + np.dot(a,c)
print([res1, res2])
```

```python
# Associative Property
n = 5
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

res1 = np.dot(a, np.dot(b, c))
res2 = np.dot(np.dot(a, b), c)
print(res1)
print(res2)
```

```python
# Special Case: When one vector is zero
n = 5
a = np.zeros(n)
b = np.random.randn(n)
c = np.random.randn(n)

res1 = np.dot(a, np.dot(b, c))
res2 = np.dot(np.dot(a, b), c)
print(res1)
print(res2)
```

```python
# Special Case: a == b == c
a = np.array([1, 3, 5, 7])
b = np.array([1, 3, 5, 7])
c = np.array([1, 3, 5, 7])

res1 = np.dot(a, np.dot(b, c))
res2 = np.dot(np.dot(a, b), c)
print(res1)
print(res2)
```

## Geometric Interpretation of Dot Product

```python
v1 = np.array([2, 4, -3])
v2 = np.array([0, -3, -3])
```

```python
# Compute the angle in radians between the two vectors
ang = np.arccos(np.dot(v1, v2)/(np.linalg.norm(v1)*np.linalg.norm(v2)))
ang
```

```python
fig = plt.figure()
ax = plt.axes(projection = "3d")
ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], "b")
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], "r")
plt.axis((-6, 6, -6, 6))
plt.title("Angle between vectors: %s rad." %ang)
plt.show()
```

```python
# The equivalence of algebraic and geometric dot product formulas
v1 = np.array([2, 4, -3])
v2 = np.array([0, -3, -3])

#algebraic
dp_a = np.dot(v1, v2)

# geometric
dp_g = np.linalg.norm(v1)*np.linalg.norm(v2)*np.cos(ang)

print(dp_a)
print(dp_g)
```

```python
# Exercise: Are these two vectors orthogonal? Yes

v1 = np.array([16, -2, 4])
v2 = np.array([0.5, 2, -1])

ang = np.arccos(np.dot(v1, v2)/(np.linalg.norm(v1)*np.linalg.norm(v2)))

cos_ang = np.cos(ang)
print(ang)
print(cos_ang)
```

```python
fig = plt.figure()
ax = plt.axes(projection = "3d")
ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], "b")
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], "r")
plt.axis((-6, 6, -6, 6))
plt.title("Angle between vectors: %s deg." %cos_ang)
plt.show()
```

## The Cauchy-Schwartz Inequality

```python
# create 3 vectors, one of which is dependent on the other
a = np.random.randn(5)
b = np.random.randn(5)
c = np.random.randn(1)*a

# Compute the dot ptoducts
aTb = np.dot(a,b)
aTc = np.dot(a,c)

# demonstrate the inequality

print(f"{np.abs(aTb):.4f}, {np.linalg.norm(a)*np.linalg.norm(b):.4f}")
print(f"{np.abs(aTc):.4f}, {np.linalg.norm(a)*np.linalg.norm(c):.4f}")
```

## Outer Product

```python
v1 = np.array([1, 2, 3])
v2 = np.array([-1, 0, 1])

# outer product
np.outer(v1, v2)
```

```python
# Using for loops
op = np.zeros((len(v1), len(v2)))
for i in range(0, len(v1)):
    for j in range(0, len(v2)):
        op[i, j] = v1[i]*v2[j]
print(op)
```

## Vector Cross Product

```python
# create two vectors
v1 = np.array([-3, 2, 5])
v2 = np.array([4, -3, 0])

# Use python's cross-product function
v3a = np.cross(v1, v2)

# Use the manual method
v3b = [ [v1[1]*v2[2] - v1[2]*v2[1]],
       [v1[2]*v2[0] - v1[0]*v2[2]],
       [v1[0]*v2[1] - v1[1]*v2[0]]]
print(v3a)
print(v3b)
```

```python
fig = plt.figure()
ax = plt.axes(projection = "3d")

# create the plane
xx, yy = np.meshgrid(np.linspace(-10, 10, 10), np.linspace(-10, 10, 10))
z1 = (-v3a[0]*xx - v3a[1]*yy)/v3a[2]
ax.plot_surface(xx, yy, z1)

# Plot the two vectors
ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], "k")
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], "k")
ax.plot([0, v3a[0]], [0, v3a[1]], [0, v3a[2]], "r")

ax.view_init(azim = 150, elev = 45)
plt.show()
```

## Unit Vector

```python
# define a vector
v1 = np.array([-3, 6])

# calculate mu
mu = 1/np.linalg.norm(v1)

# unit vector
v1n = v1*mu
np.linalg.norm(v1n)
```

```python
# Plot the stuff
plt.plot([0, v1n[0]], [0, v1n[1]], "r", label = "v1-norm", linewidth = 5)
plt.plot([0, v1[0]], [0, v1[1]], "b", label = "v1")
# square axis
plt.axis("square")
plt.axis((-6, 6, -6, 6))
plt.grid()
plt.legend()
plt.show()
```

```python
# define a vector and shrink it
v1 = np.array([-3, 6])/10

# calculate mu
mu = 1/np.linalg.norm(v1)

# unit vector
v1n = v1*mu
print(np.linalg.norm(v1))
print(np.linalg.norm(v1n))
```

```python
# Plot the stuff and see the length of unit vector being larger than the original vector
plt.plot([0, v1n[0]], [0, v1n[1]], "r", label = "v1-norm", linewidth = 5)
plt.plot([0, v1[0]], [0, v1[1]], "b", label = "v1")
# square axis
plt.axis("square")
plt.axis((-2, 2, -2, 2))
plt.grid()
plt.legend()
plt.show()
```

## Span
