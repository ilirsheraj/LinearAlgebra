# Section 6: Linear Systems of Equations

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from IPython import display
import time
# The coefficients of equations are defined as 
# ay = bx + c

eq1o = np.array([1, 2, 1])
eq2o = np.array([2, 1, 3])

# create a loop
for i in range(10):
    
    # clear the plot
    plt.cla()
    
    # randomly update equations
    eq1 = np.add(eq2o, np.random.randn(1)*eq1o)
    eq2 = np.add(eq1o, np.random.randn(1)*eq2o)
    
    # plot the new lines (solutions)
    y = ([eq1[1]*-3, eq1[1]*3] + eq1[2])/eq1[0]
    plt.plot([-3, 3], y)
    
    y = ([eq2[1]*-3, eq2[1]*3] + eq2[2])/eq2[0]
    plt.plot([-3, 3], y)    
    plt.axis([-3, 3, -3, 6])
    
    # pause to allow inspection
    display.display(plt.gcf())
    display.clear_output(wait=True)
    time.sleep(1)
```

```python
# The equation is defined as follows:
# az = bx + cy + d
eq1o = np.array([1, 2, 3, -1]) # [a, b, c, d]
eq2o = np.array([2, 1, 3, 3])

# set up the 3d plot
fig = plt.figure(figsize=(8, 8))
ax = fig.gca(projection = "3d")

for i in range(0, 15):
    
#    plt.cla()
    
    # randomly update the equations
    eq1 = np.add(eq2o, np.random.randn(1)*eq1o)
    eq2 = np.add(eq1o, np.random.randn(1)*eq2o)
    
    # plot new lines (solutions)
    y = ([eq1[1]*-3, eq1[1]*3] + eq1[2])/eq1[0]
    z = ([eq2[1]*-3, eq2[1]*3] + eq2[2])/eq2[0]
    ax.plot([-3, 3], y,z)
    
    # axis limits
    ax.set_xlim3d(-3, 6)
    ax.set_ylim3d(-3, 6)
    ax.set_zlim3d(-1, 10)
    
    # pause to allow inspection
    display.display(plt.gcf())
    display.clear_output(wait=True)
    time.sleep(.1)
```

## Reduced Row Echelon Form (RREF)

```python
# Create two random matrices
from sympy import Matrix
A = Matrix(np.random.randn(4,4))
B = Matrix(np.random.randn(4,3))
```

```python
print(np.array(A))
print()
print(np.array(B))
```

```python
# Compute RREF
rrefA = A.rref()
rrefB = B.rref()
```

```python
print(np.array(rrefA[0]))
```

```python
print(np.array(rrefB[0]))
```
